**a. Uniform Random Numbers**

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 1x10 vector of uniform random numbers
uniform_random_vector = np.random.rand(10)
print("Uniform Random Vector (1x10):\n", uniform_random_vector)

# Plot histogram
plt.hist(uniform_random_vector, bins=5, edgecolor='black')
plt.title('Histogram of Uniform Random Numbers (1x10)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**b. Normally Distributed Random Numbers**

```python
# Generate 500 random numbers from a standard normal distribution
normal_random_numbers = np.random.normal(0, 1, 500)

# Plot histogram
plt.hist(normal_random_numbers, bins=20, edgecolor='black')
plt.title('Histogram of Standard Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

**Part 1-Setup & Personalization**

```python
import random  # to create random numbers

def estimate_pi(num_points, R):
    inside = 0  # how many points inside the circle
    range_limit = R / 10

    for _ in range(num_points):
        x = random.uniform(0, range_limit)  # pick a random number from 0 to R/10
        y = random.uniform(0, range_limit)
        if x*x + y*y <= range_limit*range_limit: # Check if within the circle with radius R/10
            inside += 1  # it's inside the circle
    pi = 4 * inside / num_points
    print("Estimated pi for", num_points, "points is:", pi)

# Extract R from student ID and calculate number of letters in name
student_id = "0432220005101003"
R = int(student_id[-2:])
name = "Nusrat Jahan Rimi"
num_letters = sum(c.isalpha() for c in name)
num_points = 200 * num_letters

# Try this with the calculated number of points and R
estimate_pi(num_points, R)
```
